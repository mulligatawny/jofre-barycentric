```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('output2.txt', skiprows=1)
```

```python
# coordinates on barycentric triangle
yPlus = data[:,1]
x = data[:,3]
y = data[:,4]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.axis('equal')
cm = plt.cm.get_cmap('viridis')
sc = ax.scatter(x,y,c=yPlus,cmap=cm) # color by y+ 
ax.plot([1, 0.5], [0, 0.866], 'k-')
ax.plot([0, 0.5], [0, 0.866], 'k-')
ax.plot([0, 1], [0, 0], 'k-')
ax.plot([0.5, 1.01, 0.1], [-0.01, 0.5, 0.88], 'w.') #base boundary
ax.text(0-.05,0-.05,'1c')
ax.text(1,0-.05,'2c')
ax.text(0.5-.025,0.866+.025,'3c')
ax.patch.set_visible(False)
ax.axis('off')
cbar = plt.colorbar(sc)
cbar.set_label('y-distance')
plt.show()
```
